````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

````

````python
df=pd.read_csv('netflix_titles.csv')
````

````python
df
````

````python
df.head()
````

````python
df.info()
````

````python
df.describe()
````

````python
df["release_year"] = pd.to_datetime(df["release_year"], format = "%Y")
df["date_added"] = pd.to_datetime(df["date_added"])
df["release_year"] = df["release_year"].dt.year
df.head()
````

````python
df.shape
````

````python
df.isnull().sum()
````

````python
# calculating missing values ratio
df.isnull().sum()/len(df)*100
````

````python
#missing values ratio in director is high so dropping that column
df1=df
df.drop('director',axis=1,inplace=True)
df.head()
````

````python
df.isnull().sum()
````

````python
# replacing missing values of date added with unknown
df.date_added.fillna('unknown', inplace=True)
````

````python
df.isnull().sum()
````

````python
df.country.value_counts()
````

````python
#replacing null countries value with most occuring country
df.country.fillna("United_states",inplace= True)
````

````python
# Most occurred rating
df.rating.value_counts()
````

````python
# replacing rating with most occured rating
df.rating.fillna("TV-MA",inplace= True)
````

````python
df.isnull().sum()
````

````python
x=df.listed_in.str.split(",")
y=np.array(x)
````

````python
df.cast.fillna("unknown",inplace= True)
````

````python
#creating bar graph to understand which countries have the highest usage of netflix shows and movies
x= df.country.value_counts().head(20)
plt.figure(figsize=(15,10))
sns.set_style("whitegrid")   #set style to whitegrid
ax= sns.barplot(x.values,x.index)   #using bar for visualization
ax.set_xlabel("Number of content")
ax.set_ylabel("Number of country")
````

**TODO** - Explore one or more columns by plotting a graph below, and add some explanation about it

````python
#showing genres most shows have or we can say most common genres
x=list(set(sum(df['listed_in'].str.split(','),[])))
counts=[]
for genre in x:
    counts.append(len(df[df['listed_in'].str.contains(genre)])) 
    
tv_genre_count = pd.DataFrame([x, counts]).transpose()

tv_genre_count.columns = ['genre', 'count']
tv_genre_count = tv_genre_count.sort_values(by = ['count'], axis = 0, ascending = False)
tv_genre_count = tv_genre_count.drop(tv_genre_count.index[[0,1]])
print(tv_genre_count)
x = tv_genre_count.head(50)
g = sns.barplot(x = x['count'], y = x['genre'])

````
